```python
#import necessary modules, set up the plotting
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import matplotlib 
from matplotlib import pyplot as plt 
import GPy
```

```python
nDays = 365
x = range(0, nDays)

# year 1
x1 = 4*np.divide(x, nDays)
y1 = (np.power((3*x1 -1/6), 3) - np.power((10*x1 - 1/10), 2)  - x1)/30 + 70
y1 = y1 + np.random.normal(0, 1, nDays)

# year 2
x2 = x1 + 4
y2 = -np.power((x2 - 4), 2) + np.power(3*x2 , 1) + 63
y2 = y2 + np.random.normal(0, 1, nDays)

# year 3
x3 = x2 + 4
y3 = np.piecewise(x3, [x3 < 11, x3 > 11], [lambda x: x + 68, lambda x: -4*x + 123])
y3 = y3 + np.random.normal(0, 1, nDays)
```

```python
x = np.concatenate((x1, x2, x3))
y = np.concatenate((y1, y2, y3))
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, '+', color = 'black')
a = 0.3
ax.axvspan(0, 2, alpha=a, color='green')
ax.axvspan(3, 5, alpha=a, color='red')
ax.axvspan(8, 11, alpha=a, color='red')
ax.axvspan(6, 8, alpha=a, color='blue')
ax.axvspan(11, 12, alpha=a, color='m')

# add colored legend
plt.show()
```

```python
# Formatting the data. Obs. have to be col vectors. 
X = np.array(x, ndmin=2).transpose()
Y = np.array(y, ndmin=2).transpose()
```

```python
kernel = GPy.kern.RBF(input_dim=1, variance=1., lengthscale=3)
# http://gpy.readthedocs.io/en/deploy/GPy.core.html // Normalize doc
m = GPy.models.GPRegression(X,Y,kernel, normalizer=True)
m
```

```python
# Usual kernel and no optim
m.plot().set_ylim(50, 90)
```

```python
m.optimize_restarts(num_restarts = 3)
```

```python
# Usual kernel
display(m)
fig = m.plot().set_ylim(50, 85)
```
