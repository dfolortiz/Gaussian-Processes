```python
#import necessary modules, set up the plotting
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import matplotlib
from matplotlib import pyplot as plt
import GPy
```

# Year 1 data

```python
nDays = 365
x1 = range(0, nDays)
x1 = 4*np.divide(x1, nDays)
y1 = (np.power((3*x1 -1/6), 3) - np.power((10*x1 - 1/10), 2)  - x1)/30 + 70

plt.plot(x1, y1)
```

```python
y1 = y1 + np.random.normal(0, 1, nDays)

plt.plot(x1, y1, 'ro')
```

# Year 2 data

```python
#-(x-4)^2 + 3*x + 1
nDays = 365
x2 = range(0, nDays)
x2 = 4*np.divide(x2, nDays) + 4
y2 = -np.power((x2 - 4), 2) + np.power(3*x2 , 1) + 63

plt.plot(x2, y2)
```

```python
y2 = y2 + np.random.normal(0, 1, nDays)

plt.plot(x2, y2, 'ro')
```

# Year 3 data

```python
nDays = 365
x3 = range(0, nDays)
x3 = 4*np.divide(x3, nDays) + 8
y3 = np.piecewise(x3, [x3 < 11, x3 > 11], [lambda x: x + 68, lambda x: -4*x + 123])

plt.plot(x3, y3)
```

```python
y3 = y3 + np.random.normal(0, 1, nDays)

plt.plot(x3, y3, 'ro')
```

# All the data

```python
x = np.concatenate((x1, x2, x3))
y = np.concatenate((y1, y2, y3))

plt.plot(x, y, 'ro')
```
